```python
from pyannote.audio import Pipeline
pipeline = Pipeline.from_pretrained("pyannote/speaker-diarization")
from pydub import AudioSegment

def Split_speaker(audio_file = 'Audio_Data/123.wav'):
  """Return:
     
  """
  Audio = AudioSegment.from_file(audio_file)
  name = 'Split_speaker/' + audio_file.split('/')[-1].replace('.wav','')
  hashDict = {}
  
  diarization = pipeline(audio_file)
  i = 0
  for segment,track,label in diarization.itertracks(yield_label = True):
    Audio_speaker_i = Audio[segment.start * 1000 : segment.end * 1000]
    Audio_speaker_i.export(name + str(i) + '.wav',format = 'wav')
    hashDict[name + str(i) + '.wav'] = label
    i += 1
  return hashDict
```
